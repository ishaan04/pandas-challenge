## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#check the maximum age of the players to determine the max bin limit after it extends to 100
purchase_data_df["Age"].max()
```

```python
#creating the bins for ages '4 years (i.e. &lt;10, 10-14, 15-19, etc.)''
age_bin = [0,4,14,19,24,29,34,39,44,120]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34", "35-39", "40-44", "45+"]
```

```python
#categorize players as per age_group using pd.cut()

genders_sort["Age Group Category"] = pd.cut(genders_sort["Age"], age_bin, labels=age_groups, include_lowest=True)
genders_sort
```

```python
#calculate the total inside each age group
age_group_category = genders_sort.groupby(["Age Group Category"])
age_group_count = age_group_category["SN"].count()
age_group_count
```

```python
#convert age group category into a DF and output
age_group_count = pd.DataFrame(age_group_count)

age_group_count = age_group_count.rename(columns = {"SN" : "Total Players in Age Group"})
total_players = age_group_count

total_players["Total % of players in age group"] = total_players/player_numbers
total_players["Total % of players in age group"] = total_players["Total % of players in age group"].map('{:,.2%}'.format)
total_players
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#top 5 spenders in the game by SN
top_SN = purchase_data_df.groupby("SN")

#purchase count 
purchases_count = top_SN["Purchase ID"].count()

#total spent
total_spent = top_SN["Price"].sum()
```

```python
#merge the data sets
spenders_merge = pd.merge(purchases_count, total_spent, on="SN", how="outer")
spenders_merge
```

```python
#format the dataset
spenders_merge = spenders_merge.sort_values(["Price"], ascending=False)
spenders_merge
```

```python
#rename the columns 
spenders_merge = spenders_merge.rename(columns = {"Purchase ID": "Purchase Count", "Price":"Total Spent"})
spenders_merge
```

```python
#calculating average purchase price and add the value on the dataframe
spenders_merge["Average Purchase Price"] = spenders_merge["Total Spent"] / spenders_merge["Purchase Count"]

#format decimals
spenders_merge["Average Purchase Price"] = spenders_merge["Average Purchase Price"].map('${:,.2f}'.format)
spenders_merge
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#the most popular items based on item ID - sort the Item ID column
item_id_group = purchase_data_df.groupby("Item ID")
item_id_group.head()
```

```python
item_id_count = item_id_group["Purchase ID"].count()
```

```python
#Identify the total amount spent
item_id_sum = item_id_group["Price"].sum()
```

```python
#merge the count and sum into one table
profitable_items = pd.merge(item_id_count,item_id_sum, on="Item ID", how="outer")
```

```python
#sort the df by frequency and format columns
profitable_items = profitable_items.sort_values(["Purchase ID"], ascending = False)
profitable_items.head()

```

```python
profitable_items = profitable_items.rename(columns = {"Purchase ID": "Purchase Count", "Price": "Total Purchase Value"})
profitable_items.head()
```

```python
#add item price to the dataframe
profitable_items["Item Price"] = profitable_items["Total Purchase Value"] / profitable_items["Purchase Count"]

#locating the item id as per item name
item_name = purchase_data_df.loc[:,("Item ID", "Item Name")]

```

```python
#merge item name into the summmary df
merged_summary = pd.merge(profitable_items, item_name, on="Item ID", how="left")
merged_summary
```

```python
#the df has duplicate values, need to remove any duplicates
merged_summary = merged_summary.drop_duplicates(subset=["Item ID"])
merged_summary = merged_summary.set_index("Item ID")
merged_summary
```

```python
#format Item Price column decimals and add dollar sign
merged_summary["Item Price"]=merged_summary["Item Price"].map('${:,.2f}'.format)
merged_summary
```
